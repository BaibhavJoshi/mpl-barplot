```python
from matplotlib import pyplot as plt
import random
import numpy as np
import pandas as pd
from collections import Counter
```

```python
plt.style.use("fivethirtyeight")
```

```python
ages_x = [i for i in range(25, 36)]

print(ages_x)
```

```python
dev_y = []

for i in range(0, len(ages_x)):

    dev_y.append(round(random.uniform(35000, 80000), 2))

dev_y.sort()

print(dev_y)
```

```python
py_dev_y = []

for i in range(0, len(ages_x)):

    py_dev_y.append(round(random.uniform(45000, 90000), 2))

py_dev_y.sort()

print(py_dev_y)
```

```python
js_dev_y = []

for i in range(0, len(ages_x)):

    js_dev_y.append(round(random.uniform(45000, 90000), 2))

js_dev_y.sort()

print(js_dev_y)
```

```python
plt.figure(figsize=(8,8))

plt.bar(ages_x, dev_y, color="r", label="All devs")
plt.plot(ages_x, py_dev_y, color="b", linewidth="2", marker="o", label="Py devs")
plt.plot(ages_x, js_dev_y, color="g", linewidth="2", marker="o", label="JS devs")

plt.xlabel("Age")
plt.ylabel("Median Salary")
plt.legend()
plt.title("Median salary($) by age")
plt.show()
```

```python
x_indices = np.arange(len(ages_x))
x_width = 0.25

plt.figure(figsize=(8,8))

plt.bar(x_indices - x_width, dev_y, width=x_width, color="r", label="All devs")
plt.bar(x_indices, py_dev_y, width=x_width, color="b", label="Py devs")
plt.bar(x_indices + x_width, js_dev_y, width=x_width,  color="g", label="JS devs")

plt.xlabel("Ages")
plt.ylabel("Median Salary")
plt.title("Median salary($) by age")
plt.xticks(ticks=x_indices, labels=ages_x)
plt.legend()

plt.show()
```

```python
raw_df = pd.read_csv("./data.csv")

raw_df.head()
```

```python
c = Counter()

raw_df["LanguagesWorkedWith"].map(lambda x : c.update(x.split(";")))

print(c)
```

```python
print(c.most_common(10))
```

```python
languages_x = []
users_y = []

for item in c.most_common(15):

    languages_x.append(item[0])
    users_y.append(item[1])

print("Languages : ", languages_x)
print("Users : ", users_y)
```

```python
languages_x[11] = "Other(s)"

print(languages_x)
```

```python
x_width = 0.5

plt.figure(figsize=(10,6))

plt.barh(languages_x, users_y, color="c")

plt.title("Most popular programming languages")
plt.xlabel("Number of users")

plt.tight_layout()
plt.show()
```

```python
languages_x.reverse()
users_y.reverse()

x_width = 0.5

plt.figure(figsize=(10,6))

plt.barh(languages_x, users_y, color="g")

plt.title("Most popular programming languages")
plt.xlabel("Number of users")

plt.tight_layout()
plt.show()
```
